```python
import pandas as pd
from pathlib import Path

file_paths = [
    "./Projet+Python_Dataset_Edstats_csv/EdStatsData.csv",
    "./Projet+Python_Dataset_Edstats_csv/EdStatsCountry.csv",
    "./Projet+Python_Dataset_Edstats_csv/EdStatsSeries.csv",
    "./Projet+Python_Dataset_Edstats_csv/EdStatsFootNote.csv",
    "./Projet+Python_Dataset_Edstats_csv/EdStatsCountry-Series.csv"
]

def analyze_dataset(file_path, missing_threshold=0.9):
    print(f"--- Analyzing: {Path(file_path).name} ---")

    df = pd.read_csv(file_path)

    print(f"Shape: {df.shape[0]:,} rows × {df.shape[1]} columns")
    print("Example row:")
    print(df.columns)

    duplicate_count = df.duplicated().sum()
    print(f"{duplicate_count:,} duplicate rows.")
    if duplicate_count > 0:
        df = df.drop_duplicates().copy()

    print("Missing values per column (count and ratio):")
    missing = df.isnull().sum().to_frame("missing_count")
    missing["missing_ratio"] = missing["missing_count"] / len(df)
    print(missing.sort_values("missing_ratio", ascending=False))

    unusable_cols = missing[missing["missing_ratio"] > missing_threshold].index
    if not unusable_cols.empty:
        print(f"Dropping {len(unusable_cols)} unusable column(s) with > {missing_threshold*100:.0f}% missing values:")
        print(list(unusable_cols))
        df = df.drop(columns=unusable_cols)

    numeric_cols = df.select_dtypes(include='number').columns
    if not numeric_cols.empty:
        print("Summary statistics for numerical columns:")
        print(df[numeric_cols].describe())

    categorical_cols = df.select_dtypes(exclude='number').columns
    for col in categorical_cols:
        print(f"\nTop values in '{col}':")
        print(df[col].value_counts(dropna=False).head(3).to_string())

# Run analysis
for file_path in file_paths:
    analyze_dataset(file_path)
```

```python
import pandas as pd

country_df = pd.read_csv("./Projet+Python_Dataset_Edstats_csv/EdStatsCountry.csv")

# Step 1: Define critical columns to check for missing data
critical_columns = ["Country Code", "Short Name", "Region", "Income Group"]

# Step 2: Identify rows with missing values in critical columns
missing_mask = country_df[critical_columns].isnull().any(axis=1)

# Step 3: Identify invalid country codes (not 2–3 uppercase letters)
invalid_code_mask = ~country_df["Country Code"].str.contains(r"^[A-Z]{2,3}$", na=False)

# Step 4: Identify suspicious names (e.g., "Aggregate", "World", "Total")
suspicious_name_mask = country_df["Short Name"].str.contains(r"Aggregate|World|Region|Total", case=False, na=False)

# Step 5: Combine all masks to flag fake entries
fake_mask = missing_mask | invalid_code_mask | suspicious_name_mask

# Step 6: Extract fake entries
fake_countries = country_df[fake_mask]
print("Fake countries:")
print(fake_countries[["Country Code", "Short Name"]])

# Step 7: Filter the Country dataframe
cleaned_country_df = country_df[~country_df["Country Code"].isin(fake_countries["Country Code"])]

# Method 8: Inner join with cleaned_country_df
cleaned_series_df = pd.merge(cleaned_country_df, country_df, on="Country Code", how="inner")
```
